```python
#importing the dataset again
dataset = pd.read_csv("diabetes (2).csv")
print(dataset.shape)
dataset.head(20)
```

```python
#Mean Values
mean_bloodpressure = dataset['BloodPressure'].mean(skipna=True)
mean_BMI = dataset['BMI'].mean(skipna=True)
mean_skin = dataset['SkinThickness'].mean(skipna=True)
mean_Insulin = dataset['Insulin'].mean(skipna=True)
print ("Mean BP",mean_bloodpressure)
print ("Mean SkinThickness",mean_skin)
print ("Mean BMI",mean_BMI)
print ("Mean Insulin",mean_Insulin)
```

```python
dataset=dataset.replace({'BloodPressure': {0: mean_bloodpressure}})
dataset=dataset.replace({'SkinThickness': {0: mean_skin}})
dataset=dataset.replace({'BMI': {0: mean_BMI}})
dataset=dataset.replace({'Insulin': {0: mean_Insulin}})
print(dataset)
```

```python
# Find Correlation
correl = dataset[dataset.columns].corr()
sns.heatmap(correl, annot = True)
plt.show()
```
